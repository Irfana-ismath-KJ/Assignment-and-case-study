```python
plt.bar(data1['day'],data1['total_bill'],color='r')
plt.title('BAR CHART OF TOTAL BILL PER DAY')
plt.xlabel('Days')
plt.ylabel('total_bill')
plt.show()
```

```python
plt.bar(data1['size'],data1['total_bill'],color='g')
plt.title('Total bill by size')
plt.xlabel('Size')
plt.ylabel('total_bill')
plt.show()
```

maximum bill generated by 3 people and minimum by 1

```python
plt.figure(figsize=(8,4))

plt.hist(data1['total_bill'],color = 'g',rwidth =0.98)

plt.title('HISTOGRAM OF TOTAL BILL',size = 15,color ='b')
plt.xlabel('TOTAL BILL')
plt.ylabel('COUNT')

plt.show()
```

maximum bill amount lies in between 10-20 and minimum lies in between 40-50

```python
plt.figure(figsize=(8,4))

plt.hist(data1['tip'],color = 'orange',rwidth =0.98)

plt.title('HISTOGRAM OF TIP',size = 15,color ='b')
plt.xlabel('TIP')
plt.ylabel('COUNT')

plt.show()
```

maximum tip amount lies in between 2-3 and minimum lies in between 8-10

```python


plt.hist(data1['size'],color = 'blue')

plt.title('HISTOGRAM OF SIZE',size = 15,color ='b')
plt.xlabel('SIZE')
plt.ylabel('COUNT')

plt.show()
```

most of the orders generated for two people and less for 1

```python
plt.figure(figsize = (8,4))

sns.countplot(x ="sex",data = data1,width = 0.5)

plt.title('COUNT PLOT OF BILL PAYER BY SEX',size = 15,color ='r')

plt.show()
```

more bill generated by male customers

```python
plt.figure(figsize = (8,4))

sns.countplot(x ="smoker",data = data1,width = 0.5)

plt.title('COUNT PLOT OF SMOKERS PRESENT',size = 15,color ='r')

plt.show()
```

majority of customers are non-smokers

```python
plt.figure(figsize = (8,4))

sns.countplot(x ="time",data = data1,width = 0.5)

plt.title('COUNT PLOT OF BILL PAYER BY TIME',size = 15,color ='r')

plt.show()
```

more people prefer for dinner than lunch

```python
sns.scatterplot(x="tip", y="total_bill", data=data1)

plt.title('SCATTER PLOT OF TOTAL BILL BY TIP',size = 15,color ='purple')

plt.show()
```
